```python
# What is the number of wines with quality 6 or greater where alochol level is less than 10 and ph level is less 5?

df[(df['quality'] >= 6) & (df['alcohol'] < 10) & (df['pH'] < 5)].shape[0]
```

```python
# draw me a bar chart of the number of wines with quality 6 or greater where alochol level is less than 10 and ph level is less 5
df[(df['quality'] >= 6) & (df['alcohol'] < 10) & (df['pH'] < 5)]['quality'].value_counts().plot(kind='bar', figsize=(10, 6), color='skyblue')
plt.title('Number of Wines with Quality 6 or Greater, Alcohol < 10, and pH < 5')
plt.xlabel('Quality')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.show()
```
